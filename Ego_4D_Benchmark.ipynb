```python
%load_ext tensorboard
import os
logdir = os.path.join(os.path.expanduser("~"),
                      "ingegneriaMatematica", "machineLearning",
                      "episodic-memory", "NLQ", "VSLNet", "runs")
%tensorboard --logdir "{logdir}"
```

```python
%%bash

BASE="/home/nicolo/ingegneriaMatematica/machineLearning"
REPO="$BASE/episodic-memory/NLQ/VSLNet"
DATA="$BASE/ego4d_data"
MODEL_SUBDIR="$REPO/runs/omnivore_vslnet"
mkdir -p "$MODEL_SUBDIR"
HSA_OVERRIDE_GFX_VERSION=11.0.0 HIP_VISIBLE_DEVICES=0 HIP_LAUNCH_BLOCKING=1
cd "$REPO"
python3 main.py \
  --model_name vslnet \
  --task nlq_official_v1_omnivore \
  --predictor bert \
  --dim 128 \
  --mode train \
  --video_feature_dim 1536 \
  --max_pos_len 128 \
  --init_lr 0.0025 \
  --epochs 10 \
  --batch_size 32 \
  --fv official \
  --num_workers 2 \
  --data_loader_workers 1 \
  --model_dir "$MODEL_SUBDIR" \
  --eval_gt_json "$DATA/v1/annotations/nlq_val.json" \
  --log_to_tensorboard "omnivore_vslnet_bs32_dim128_epoch10_ilr0.0025" \
  --tb_log_freq 5 \
  --remove_empty_queries_from train
```
